```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math 
import copy
# reading the data
df = pd.read_csv('polynomial_train.csv')
temp = df.to_numpy()
X_train = temp[:,1:4]
y_train = temp[:,4]

```

```python
def zscore(X_in):
    mu     = np.mean(X_in, axis=0)                 
    sigma  = np.std(X_in, axis=0)                
    X_norm = (X_in - mu) / sigma  
    return (X_norm, mu, sigma)
X,mu,sigma = zscore(X_train)
  
    
degree = int(input("input the degree u want: "))
for i in range (degree+1):
    for j in range (degree+1):
        for k in range(degree+1):
            if (i+j+k)>degree or (i+j+k)<=1:
                continue
            else:
                r = (X[:,0]**i)*(X[:,1]**j)*(X[:,2]**k)
                print(f"{i} and {j} and {k} ") 
                X = np.hstack((X, np.atleast_2d(r).T))
            
print(X.shape)           
```

```python
# defining the cost function
def compute_cost(x,y,w,b):
    m,n = x.shape
    cost = ((np.matmul(x,w) + b) - y)**2
    total_cost = np.sum(cost)/(2*m)
    return total_cost
```

```python
# defining the gradient descent
# defining the gradient descent
def compute_gradient(x, y, w, b): 
    
    m,n = x.shape          
    dj_dw = np.zeros((m,n))
    dj_db = np.zeros(m)
    
    c = ((np.matmul(x,w) + b) - y)
    for i in range(m):                             
        dj_dw[i] = c[i]*x[i]
    dj_dw = np.sum(dj_dw,axis = 0)/m
    dj_db = np.sum(c)/m
                                 
        
        
    return (dj_db,dj_dw)
```

```python
 # defining the gradient descent function
def gradient_descent(x,y,w_in,b_in,cost_function,gradient_function,num_iters,alpha):
    I = []
    J = []
    
    w = copy.deepcopy(w_in)
    b = b_in
    
    for i in range (num_iters):
        dj_db, dj_dw = gradient_function(x,y,w,b)
        w = w - (alpha * dj_dw)
        b = b - (alpha * dj_db)
        
        if i < 10000:
            J.append(cost_function(x,y,w,b))
            I.append(i)
        
        if i% math.ceil(num_iters/10) == 0:
            print(f"Iteration : {I[-1]} and Cost: {J[-1]} ")
        print(i)
    return w,b,J,I
# running the program
m,n = X.shape
b_init = 0 
w_init = np.zeros(n)
iteration = 1500

a = 0.0001


w_final,b_final,J_his,i_his = gradient_descent(X,y_train,w_init,b_init,compute_cost,compute_gradient,iteration,a)
            
print(f"final required w,b : {w_final} and {b_final}")

data = {
  "Cost": J_his, 
  "Iteration": i_his
}
df=pd.DataFrame(data)
df
    
```
