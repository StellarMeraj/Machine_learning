```python
#read the data 
# reading the data
df = pd.read_csv('classification_train.csv')
temp1 = df.to_numpy()
X_train = temp1[5000:30001,2:787]
X_test = temp1[1:5001,2:787]
y_train = temp1[5000:30001,0:2]
y_test = temp1[1:5001,0:2]
print(X_train.shape)
print(X_test.shape)
print (y_train.shape)
print(y_test.shape)
```

```python
def zscore(X_in):
    mu     = np.mean(X_in, axis=0)                 
    sigma  = np.std(X_in, axis=0)                
    X_norm = (X_in - mu) / sigma
    X_op = X_in/ 255
    return X_norm ,mu,sigma
a,mean,std = zscore(X_train)
print(a.shape)
```

```python
#making 10 different ytrains sets
m = y_train.shape[0]
y_ntrain = y_train[:,1:2]
for i in range(10) :
    y_new = np.zeros(m)
    for j in range(m):
        if y_train[j,1] == i:
            y_new[j] = 1
        else:
            y_new[j] = 0 
    y_ntrain = np.hstack((y_ntrain, np.atleast_2d(y_new).T))
print(y_ntrain.shape)
print(y_ntrain[:,0:11])
```

```python
# computoing the sigmoid fucntion
def compute_sigmoid(z):
    a = 1/(1+np.exp(-z))
    return a

```

```python
# computing the cost function
def compute_cost(x,y,w,b):
    m = x.shape[0]
    z_i = np.matmul(x,w) + b
    f_wb = compute_sigmoid(z_i)
    cost = -(1/m)*np.sum(y*(np.log(f_wb)) + (1-y)*(np.log(1-f_wb)))
    return cost
```

```python
def compute_gradient(x,y,w,b):
       
    m,n = x.shape          
    dj_dw = np.zeros((m,n))
    dj_db = np.zeros(m)
    c = compute_sigmoid((np.matmul(x,w) + b)) - y
    for i in range(m):                             
        dj_dw[i] = c[i]*x[i]
    dj_dw = np.sum(dj_dw,axis = 0)/m
    dj_db = np.sum(c)/m
                                 
        
        
    return (dj_dw,dj_db)
```

```python
def compute_gradient_descent(x,y,w_in,b_in,alpha,no_iters,gradient,cost):
    cost_store=[]
    w=w_in
    b = b_in
    for i in range(no_iters):
        dj_dw,dj_db = gradient(x,y,w,b)
        w = w - alpha*dj_dw
        b = b - (alpha*dj_db)
        coste=cost(x,y,w,b)
        cost_store.append(coste)
        print(coste)
    return w,b
```

```python
m,n = X_train.shape
b_init = 0
w_init = np.zeros(n)
w_outr = np.zeros((10,784))
b_outr = np.zeros((10,1))
X = zscore(X_train)
for i in range(1,11):
    no_iters = 200
    if(i==9):
        alpha=0.01
    else:
        alpha=0.05
    w_out,b_out = compute_gradient_descent(X,y_ntrain[:,i],w_init,b_init,alpha,no_iters,compute_gradient,compute_cost)
    if(no_iters%25==0):
        print(f"the final w,b for {i-1} th iteration = {w_out},{b_out} ")
    w_outr[i-1] = w_out
    b_outr[i-1] = b_out
    
```
