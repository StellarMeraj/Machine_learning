```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(50)
data=pd.read_csv('classification_train.csv')
print(data.head(10))
m=len(data)
n=len(data.columns)
data_of_each_individual_x={}
for i in range(0,(n-2)):
    data_of_each_individual_x[i]=data.iloc[:,(i+2)]
print(data_of_each_individual_x)
```

```python
print(m)
print(n)
```

```python
for i in range((n-2)):    
    data_of_each_individual_x[i]/=255
print(data_of_each_individual_x)
```

```python
data_of_all_x={}
for i in range(m):
    loop_array=data.iloc[i]
    data_of_all_x[i]=loop_array[2:n]
print(data_of_all_x[1])
```

```python
z={}
w=np.zeros((n-2))
y_hat=np.zeros(m)
b=0
for i in range(m):
    z[i]=np.dot(w,data_of_all_x[i])+b
    y_hat[i]=(1+np.exp(-z[i]))**(-1)
print(y_hat)
```

```python
individual_classification_data={}
for i in range(10):
    individual_classification_data[i]=data.iloc[:,1].copy()
```

```python
for i in range(10):
    for j in range(m):
        if(individual_classification_data[i][j]==i):
            individual_classification_data[i][j]=1
        else:
            individual_classification_data[i][j]=0
    print(individual_classification_data[i])
```

```python
cost_data={}
w_data={}
b_data={}
for i in range(10):
    for no_of_iterations in range(50):
        data_of_grad_w=0
        grad_b=(1/m)*(np.sum(y_hat-individual_classification_data[i]))
        temp_w=np.zeros((n-2))
        for j in range((n-2)):
            data_of_grad_w=((1/m)*((np.sum((y_hat-individual_classification_data[i])*data_of_each_individual_x[j]))))
            temp_w[j]=w[j]-(0.0005)*data_of_grad_w
        temp_b=b-((0.0005)*(grad_b))
        w=temp_w
        b=temp_b
        for k in range(m):
            y_hat[k]=((1+np.exp(-np.dot(w,data_of_all_x[k])+b))**(-1))
        print(no_of_iterations+1)
        cost_function=-(1/m)*(np.sum((individual_classification_data[i]*(np.log(y_hat)))+((1-individual_classification_data[i])*(np.log(1-y_hat)))))
        print(cost_function)
        if((no_of_iterations+1)==50):
            cost_data[i]=cost_function
            w_data[i]=w
            b_data[i]=b
```

```python
print(cost_data)
print(w_data)
print(b_data)
```
